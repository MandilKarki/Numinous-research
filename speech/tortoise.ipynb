```python
from TTS.api import TTS
tts = TTS("tts_models/en/multi-dataset/tortoise-v2")

```

```python
# cloning `lj` voice from `TTS/tts/utils/assets/tortoise/voices/lj`
# with custom inference settings overriding defaults.
tts.tts_to_file(text="Hello, my name is Mandil , how are you? I feel shiva is a state of mind",
                file_path="/content/output0.wav",
                voice_dir="/content/",
                num_autoregressive_samples=1,
                diffusion_iterations=10)

# Using presets with the same voice
tts.tts_to_file(text="Hello, my name is Mandil , how are you?my kundalini is awakening",
                file_path="output1.wav",
                voice_dir="/content/",
                preset="ultra_fast")

# Random voice generation
tts.tts_to_file(text="Hello, my name is Mandil , how are you?my kundalini is awakening the serpant lurks ",
                file_path="/content/output2.wav")
```
